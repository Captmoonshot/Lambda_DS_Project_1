####  But neither the histograms nor the tabular data summaries of the adjusted closing price tell us whether there was a bubble or not.  We could possibly go through the entire dataset row by row but that would be incredily tedious and boring and prone to mistakes.

#### Let's do some more visualizations:

```python
# We add a percent_change column that calculates the percentage change of adjusted closing price of the assets

sp500['percent_change'] = sp500['Adj Close'].pct_change() * 100
sp500['percent_change'] = sp500['percent_change'].bfill()
sp500.head()
```

```python
# Same procedure for Bitcoin and MoviePass

btc['percent_change'] = btc['Adj Close'].pct_change() * 100
btc['percent_change'] = btc['percent_change'].bfill()

hmny['percent_change'] = hmny['Adj Close'].pct_change() * 100
hmny['percent_change'] = hmny['percent_change'].bfill()
```

```python

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, sharey=False, figsize=(14, 8))

ax1.plot(sp500['Adj Close'])
ax1.set_title('S&P 500')
ax1.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)

ax2.plot(btc['Adj Close'], 'r')
ax2.set_title('Bitcoin')
ax2.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)

ax3.plot(sp500['percent_change'])
ax3.set_title('S&P 500 % Change')
ax3.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)

ax4.plot(btc['percent_change'], 'r')
ax4.set_title('Bitcoin % Change')
ax4.text(x='2014-04', y=280, s='339% change in price in one day', fontweight='bold', color='xkcd:crimson')
ax4.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)

f.savefig('S&P_vs_Bitcoin.png')
```

#### And here we have our first real visual confirmation of an asset bubble comparing the price of Bitcoin to the S&P 500 Index: the classic hockey-stick pattern of going up and then going down just as fast if not faster.  

#### The bottom plots show daily percentage changes.  Note the 339% increase in price that Bitcoin experienced on February 14th 2014.

#### Let's have a second look at the plots of the S&P 500 index compared to those of Bitcoin and MoviePass.

#### We're actually going to use all the data that Yahoo is willing to give us and extend our current DataFrames.  At the same time we're going to create a new column for each of our assets which will capture the change in the 'Adj Close' price feature from one period to the next. The new column will be called 'Adj_Close_Change', and by definition it's the difference between the 'Adj Close' shifted forward by one period and the current 'Adj Close' => Change in y= y<sub>t</sub> - y<sub>t-1</sub>...also known as the first-difference.

```python
# S&P 500 Index
sp500_big = web.get_data_yahoo('^GSPC', start=datetime.datetime(1980, 1, 3), end=datetime.datetime(2018, 12, 10))
sp500_big['Adj_Close_Shifted'] = sp500_big['Adj Close'].shift(1).bfill()
sp500_big['Adj_Close_Change'] = sp500_big['Adj_Close_Shifted'] - sp500_big['Adj Close']

# Bitcoin
btc_big = web.get_data_yahoo('BTC-USD', start=datetime.datetime(2010, 1, 3), end=datetime.datetime(2018, 12, 10))
btc_big['Adj_Close_Shifted'] = btc_big['Adj Close'].shift(1).bfill()
btc_big['Adj_Close_Change'] = btc_big['Adj_Close_Shifted'] - btc_big['Adj Close']

# MoviePass stock
hmny_big = web.get_data_yahoo('hmny', start=datetime.datetime(1997, 1, 3), end=datetime.datetime(2018, 12, 10))
hmny_big['Adj_Close_Shifted'] = hmny_big['Adj Close'].shift(1).bfill()
hmny_big['Adj_Close_Change'] = hmny_big['Adj_Close_Shifted'] - hmny_big['Adj Close']

sp500_big.head()
```

#### Second, if we compare the means of the 'Adj Close' price for S&P 500 from the time period (1980 - 1995) and compared it to the time period (1996 - 2018), there's a drastic difference in the magnitude.  

#### What these two observations tell us is that the time series data for S&P 500 falls into a category of time series data called non-stationary data.  

#### The two defining characteristics of non-stationary data are:
1. Non-constant means across different time periods
2. Non-constant means even when taking the difference of means from period to period, so the the change in means are also non-constant.

#### Time series data that are non-stationary due to different means across different time periods can be made stationary if the change in means are constant (fluctuate around the same point up and down for a long time). 

#### Why is this important?

#### Because stationary time series, as opposed non-stationary time series, revert to the mean and therefore can be analyzed with statistical methods for things like forecasting.

#### If you examine the adjusted closing price for Bitcoin, from 2011 to 2017, it resembles what's known as a random walk with no clear trend until you get to the time period from 2017 - 2018.

#### The same situation applies to the adjusted closing price of HMNY (MoviePass) stock from about 2003 to 2018, and resemble a random walk even though the fluctuations seem greater in magnitude.

#### As described above, a way to tell if a time series can be made stationary and therefore mean reverting is to look at the differences in means as we do here.

```python
f, axarr = plt.subplots(3, sharex=False, figsize=(12,8))
axarr[0].plot(sp500_big['Adj Close diff'], color='xkcd:azure')
axarr[0].set_title('Bitcoin and MoviePass are Mean Reverting Assets')
axarr[1].plot(btc_big['Adj Close diff'], color='xkcd:gold')
axarr[2].plot(hmny_big['Adj Close diff'], color='xkcd:crimson')

f.legend(['S&P 500', 'Bitcoin', 'MoviePass']);
f.savefig('S&P500_Bitcoin_HMNY_Mean_Reversion.png')
```

#### As the above plots show the means for S&P 500 Index is clearly non-stationary as even the change in means has an increasing trend.

#### Also, there appears to be some strong evidence that both Bitcoin and MoviePass stock are stationary and therefore mean reverting for most of their price histories.  

#### Thus, for the average investor, it may not be prudent to invest in an asset that has:
1. Long-run mean-reverting properties and
2. Goes up in price to magnitudes previously unseen (hockey-stick pattern) without good reasons and 
3. Highy volatile and
4. The investor has no plans to hold the asset for the long-term
