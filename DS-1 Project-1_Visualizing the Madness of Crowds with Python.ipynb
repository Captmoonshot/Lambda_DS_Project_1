# Visualizing the Madness of Crowds with Python

#### For this project we're going to look at some exemplary financial time series to get a handle on some of the salient characteristics of asset bubbles.

#### We're also going to use a new library in addition to the standard libraries we use for data analysis:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# New library
import pandas_datareader.data as web

import datetime

```

#### For analyzing financial assets it's standard practice in the industry to compare an investment performance against some benchmark such as the S&P 500 which is an index made up of 500 large market-cap companies in the U.S. listed on the NYSE and the Nasdaq.   Our plan of action will be to compare two assets (Bitcoin and Helios & Matheson stock) with a history of known bubble-like episodes against the S&P 500 benchmark in a visual fashion.  

#### Here, we'll be creating a pandas DataFrame for the S&P 500 along with datasets for Bitcoin (the cryptocurrency) and Helios and Matheson which is the public company behind Movie Pass.  And the main variable of interest for our DataFrames is 'Adjusted Close' which is the regular price adjusted for events like stock splits and therefore a more accurate reflection.

#### Our DataFrames:
1. sp500 (S&P 500 index)
2. btc (Bitcoin)
3. hmny (Helios & Matheson stock)

#### We're also making sure all the variables we examine are in the same time frame, in this case from the year 2013 to 2018.  So all of our DataFrames will be in that time frame as their respective indexes with daily frequencies as the frequency.

```python
# S&P 500 DataFrame

sp500 = web.get_data_yahoo('^GSPC', start=datetime.datetime(2013, 1, 3), end=datetime.datetime(2018, 12, 10))
sp500.head()
```

```python
sp500.shape
```

```python
# Bitcoin DataFrame

btc = web.get_data_yahoo('BTC-USD', start=datetime.datetime(2013, 1, 3), end=datetime.datetime(2018, 12, 10))
btc.head()                         
```

```python
# Helios and Matheson (HMNY) DataFrame

hmny = web.get_data_yahoo('HMNY', start=datetime.datetime(2013, 1, 3), end=datetime.datetime(2018, 12, 10))
hmny.head()
```

#### The first things we're going to do is examine the data is tabular format to see if can glean some insights from it:

```python
sp500['Adj Close'].describe()
```

```python
btc['Adj Close'].describe()
```

```python
hmny['Adj Close'].describe()
```

#### From looking at the mean, median, and standard deviation of the data, we know:

1. S&P500 seems normal with similar mean and median, and a standard deviation that is about 17% of the mean.  
2. Bitcoin has a mean that's about 4x the median which suggests a right skew, and a standard deviation that is much larger than the mean.
3. Similarly to Bitcoin, HMNY (MoviePass stock) has a mean that's larger than the median which suggests a right skew, and a standard deviation that's almost as large as the mean.

#### We can confirm our intuitions by doing some simple histograms:

```python

plt.style.use('fivethirtyeight')

f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=False, sharey=False, figsize=(14, 8))
ax1.hist(sp500['Adj Close'], color='xkcd:blue')
ax1.set_title('S&P 500 Histogram')
ax2.hist(btc['Adj Close'], color='xkcd:gold')
ax2.set_title('Bitcoin Histogram', color='xkcd:gold')
ax3.hist(sp500['Adj Close'])
ax3.set_title('S&P 500 Histogram')
ax4.hist(hmny['Adj Close'], color='xkcd:crimson')
ax4.set_title('HMNY "MoviePass" Histogram', color='xkcd:crimson');
f.savefig('Histograms.png')
```

#### The histograms confirm that both Bitcoin and HMNY (MoviePass) have skewed distributions relative to the somewhat normal distribution of the S&P 500.

####  But neither the histograms nor the tabular data summaries of the adjusted closing price tell us whether there was a bubble or not.  We could possibly go through the entire dataset row by row but that would be incredily tedious and boring and prone to mistakes.

#### Let's do some more visualizations:

```python
# We add a percent_change column that calculates the percentage change of adjusted closing price of the assets

sp500['percent_change'] = sp500['Adj Close'].pct_change() * 100
sp500['percent_change'] = sp500['percent_change'].bfill()
sp500.head()
```

```python
# Same procedure for Bitcoin and MoviePass

btc['percent_change'] = btc['Adj Close'].pct_change() * 100
btc['percent_change'] = btc['percent_change'].bfill()

hmny['percent_change'] = hmny['Adj Close'].pct_change() * 100
hmny['percent_change'] = hmny['percent_change'].bfill()
```

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex=True, sharey=False, figsize=(14, 8))
ax1.plot(sp500['Adj Close'])
ax1.set_title('S&P 500')
ax1.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax2.plot(btc['Adj Close'], 'r')
ax2.set_title('Bitcoin')
ax2.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax3.plot(sp500['percent_change'])
ax3.set_title('S&P 500 % Change')
ax3.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax4.plot(btc['percent_change'], 'r')
ax4.set_title('Bitcoin % Change')
ax4.text(x='2014-04', y=280, s='339% change in price in one day', fontweight='bold', color='xkcd:crimson')
ax4.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
f.savefig('S&P_vs_Bitcoin.png')
```

#### And here we have our first real visual confirmation of an asset bubble comparing the price of Bitcoin to the S&P 500 Index: the classic hockey-stick pattern of going up and then going down just as fast if not faster.  

#### The bottom plots show daily percentage changes.  Note the 339% increase in price that Bitcoin experienced on February 14th 2014.  

#### Next we look at MoviePass stock:

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True, sharey=False, figsize=(14, 8))
ax1.plot(sp500['Adj Close'])
ax1.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax1.set_title('S&P 500')
ax2.plot(hmny['Adj Close'], 'r')
ax2.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax2.set_title('HMNY "MoviePass"')
ax3.plot(sp500['percent_change'])
ax3.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax3.set_title('S&P 500 % Change')
ax4.plot(hmny['percent_change'], 'r')
ax4.set_title('HMNY "MoviePass" % Change')
ax4.text(x='2016-08-01', y=500, s='900% change in price', fontweight='bold', color='xkcd:crimson')
ax4.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
f.savefig('S&P_vs_HMNY.png')
```

#### Again, we see the same hockey-stick pattern, but this time MoviePass stock shows at least two different bubble-like price action, with an absolutely astounding 900% increase in price on June 7th 2016.  

#### While tabular data did give us succint summaries of the datasets, and some important clues in the mean, median, min, and max numbers, only by visualizing can we get a full sense of the drama involved in the dynamics of price changes.  

#### Oftentimes, to get a point across to the population at large, a high-impact visual can do more with one plot than a thousand academic papers only a few will understand.  

#### Let's have a look a the volatility of the assets:

```python
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, sharex=True, sharey=True, figsize=(14, 8))
ax1.plot(sp500['percent_change'])
ax1.set_title('S&P 500 % Change')
ax1.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax2.plot(hmny['percent_change'], 'r')
ax2.set_title('HMNY "MoviePass" % Change')
ax2.text(x='2016-08-01', y=500, s='900% change in price', fontweight='bold', color='xkcd:crimson')
ax2.text(x='2016-08-01', y=450, s='in two days', fontweight='bold', color='xkcd:crimson')
ax2.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax3.plot(sp500['percent_change'])
ax3.set_title('S&P 500 % Change')
ax3.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
ax4.plot(btc['percent_change'], 'r')
ax4.set_title('Bitcoin % Change')
ax4.text(x='2014-04', y=300, s='339% change in price in one day', fontweight='bold', color='xkcd:crimson')
ax4.axhline(y=0, color='black', linewidth=1.3, alpha=0.8)
f.savefig('S&P500vsHMNYvsBitcoin')
```

#### The above plots shows price changes and gives us strong clues as to their volatility.  Scaled with the same y-axis, they show that Bitcoin and MoviePass stocks may be higher-than-average risk assets than the average investor might want to stay away from.
